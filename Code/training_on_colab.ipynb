```python
from pathlib import Path
```

```python
TRAJECTORY_DUR = 1000


checkpoint_path = "/content/drive/My Drive/DeepLearningTP5/Code/wandb/run-20210305_035512-zkzrkghd/files/test_rossler/zkzrkghd/checkpoints/epoch=4-step=49999.ckpt"


trained_model = Model.load_from_checkpoint(checkpoint_path=checkpoint_path)
trained_model.normalize = False
nb_steps = int(TRAJECTORY_DUR // trained_model.delta_t)

rossler_map_true = RosslerMap(delta_t=trained_model.delta_t)

# Train set
traj_pred, traj_true, time_list = compute_pred_true_traj(
    trained_model, rossler_map_true, [-5.75, -1.6, 0.02], nb_steps
)
```

```python
checkpoint_path = Path(checkpoint_path)
save_dir_path = checkpoint_path.parent

np.save(os.path.join(save_dir_path, "traj_pred_train.npy"), traj_pred)
np.save(os.path.join(save_dir_path, "traj_true_train.npy"), traj_true)
np.save(os.path.join(save_dir_path, "time_list_train.npy"), time_list)
```

```python
plot_pred_true_trajectories(None, traj_pred, traj_true, time_list, prefix="train_")
compute_pred_true_lyaponov(
    None,
    trained_model,
    rossler_map_true,
    traj_pred,
    traj_true,
    nb_steps,
    mode="discrete",
)
compute_pred_true_equilibrium_state(None, trained_model, rossler_map_true, mode="discrete")
```
